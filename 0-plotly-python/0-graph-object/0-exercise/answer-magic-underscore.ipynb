> Using the data above, create an interactive vertical bar chart with Plotly that: <br>
> 1. Displays **Sales by Category** as the chart title
> 2. Labels the x-axis ```"Category"``` and y-axis ```"Sales (USD)"```
> 3. Applies the custom color for each bar (use the ```colors``` list)
> 4. Adds **text labels on top of each bar** showing the sales value
> 5. Uses a white background (use ```"plotly_white``` template)

````python
import plotly.graph_objects as go

categories = ["Electronics", "Clothing", "Food", "Books", "Toys"]
sales = [45200, 31800, 27500, 15400, 19700]
colors = ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]

fig = go.Figure(
    data=[
        go.Bar(
            x=categories,
            y=sales,
            marker_color=colors,
            text=sales,
            textposition="outside"
        )
    ]
)

fig.update_layout(
    title="Sales by Category",
    xaxis_title="Category",
    yaxis_title="Sales (USD)",
    template="plotly_white"
)

fig.show()
````
