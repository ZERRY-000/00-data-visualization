### Exercise

Create an **interactive bar chart** using `plotly.express` that:

1. x-axis is `"Category"`, y-axis is `"Sales"`
2. Color each bar by `"Region"`
3. Title: `"Sales by Category"`
4. Label x-axis `"Category"`, y-axis `"Sales (USD)"` using `labels` parameter
5. Use template `"plotly_white"`

> Hint: Check [`plotly.com/python/bar-charts/`](https://plotly.com/python/bar-charts/) — 
> Plotly Express uses `px.bar()` and most things go **directly as parameters**, 
> no `update_traces` needed!

```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame({
    "Category": ["Electronics", "Clothing", "Food", "Books", "Toys"],
    "Sales": [45200, 31800, 27500, 15400, 19700],
    "Region": ["North", "South", "North", "East", "West"]
})

fig = px.bar(
    df,
    x="Category",
    y="Sales",
    color="Region",
    title="Sales by Category",
    labels={"Sales": "Sales (USD)", "Category": "Category"},
    template="plotly_white"
)

fig.show()
```
